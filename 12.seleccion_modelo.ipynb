##Medianos

```python
import os

columns_resultados = ['NOMBRE', 'CONFIG', 'R2_MODEL', 'CROSS_ECM', 'CROSS_ECM_DE', 'CROSS_R2', 'CROSS_R2_DE', 'VARIANZA_EXPL', 'MAE', 'MAE_POISSON', 'MAE_GAMMA']
columns_resultados_con_modelo = ['MODEL','NOMBRE', 'CONFIG', 'R2_MODEL', 'CROSS_ECM', 'CROSS_ECM_DE', 'CROSS_R2', 'CROSS_R2_DE', 'VARIANZA_EXPL', 'MAE', 'MAE_POISSON', 'MAE_GAMMA']
```

```python
df_resultados_medianos_mejores_3 = pd.DataFrame(columns=columns_resultados_con_modelo, index=None)
df_resultados_medianos_todos = pd.DataFrame(columns=columns_resultados_con_modelo, index=None)

for filename in os.listdir(RUTA_RESULTADOS_MEDIANOS):
    if filename.endswith("csv"): 
      print(filename)
      url = RUTA_RESULTADOS_MEDIANOS + '/' + filename
      df_todos = pd.read_csv(url, header = None, index_col= False, names = columns_resultados)
      df_todos['MODEL'] = filename;
      df_mejores_3 = df_todos.sort_values(by='MAE').head(3)

      df_resultados_medianos_mejores_3 = pd.concat([df_resultados_medianos_mejores_3, df_mejores_3]);
      df_resultados_medianos_todos = pd.concat([df_resultados_medianos_todos, df_todos]);
```

### Teniendo en cuenta los 3 mejores de cada modelo

```python
df_resultados_medianos_mejores_3.sort_values(by='MAE')
```

### Teniendo en cuenta todos los resultados de cada modelo

```python
pd.set_option("min_rows", 20)
df_resultados_medianos_todos.sort_values(by='MAE', ascending=True)
```

### Mejor modelo

```python
df_resultados_medianos_todos.sort_values(by='MAE', ascending=True).head(1)
```

###Peor modelo

```python
df_resultados_medianos_todos.sort_values(by='MAE', ascending=False).head(1).round(1)
```
